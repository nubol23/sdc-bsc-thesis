```python
dataset = pd.read_csv('/home/nubol23/Documents/DriveDatasetStable/Val/val_dataset.csv')
junctions = pd.read_csv('/home/nubol23/Documents/DriveDatasetStable/Val/juncs.csv')

# dataset = pd.read_csv('/home/nubol23/Documents/DriveDatasetStable/Train/train_dataset.csv')
# junctions = pd.read_csv('/home/nubol23/Documents/DriveDatasetStable/Train/juncs.csv')
```

```python
dataset.head()
```

```python
dataset['folder'] = [f.split('/')[0] for f in dataset['filenames'].tolist()]
dataset['file'] = [f.split('/')[1].split('.')[0] for f in dataset['filenames'].tolist()]
dataset['prev_throttle'] = [0] + dataset['throttle'].tolist()[:-1]

dataset = dataset.drop(['filenames'], axis=1)

dataset.head()
```

```python
junctions.head()
```

```python
path_left = [0]*len(dataset)
path_right = [0]*len(dataset)
path_forward = [0]*len(dataset)

action_left = [0]*len(dataset)
action_right = [0]*len(dataset)
action_forward = [0]*len(dataset)
no_action = [1]*len(dataset)
```

```python
import os
```

```python
junc_path = '/home/nubol23/Documents/DriveDatasetStable/Val/junctions'
# junc_path = '/home/nubol23/Documents/DriveDatasetStable/Train/junctions'
juncs = sorted(os.listdir(junc_path), key=lambda x: int(x))

for junc in juncs:
    files = sorted(os.listdir(f'{junc_path}/{junc}'),
                   key=lambda x: int(x.split('.')[0].split('_')[1]))
    
    row = junctions.iloc[int(junc)]
    for file in files:
        folder, img = list(map(int, file.split('.')[0].split('_')))
        idx = 8000*folder + img
        
        path_left[idx] = row['path_left']
        path_right[idx] = row['path_right']
        path_forward[idx] = row['path_forward']

        action_left[idx] = row['action_left']
        action_right[idx] = row['action_right']
        action_forward[idx] = row['action_forward']
        no_action[idx] = 0
    #18229
```

```python
res = dataset.copy()

res['path_left'] = path_left
res['path_right'] = path_right
res['path_forward'] = path_forward

res['action_left'] = action_left
res['action_right'] = action_right
res['action_forward'] = action_forward
res['no_action'] = no_action

res.head()
```
